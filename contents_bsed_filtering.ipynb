```
import pandas as pd
import numpy as np
```

```
user_table = pd.read_excel('/content/drive/My Drive/프로젝트/트래블 메이커/Data/User Table_10000.xlsx', index_col=0)
user_table = user_table.reset_index()
user_table = user_table.drop(['Rank', 'Rand'], axis=1)
user_table.head()
```

```
item_table = pd.read_csv('/content/drive/My Drive/프로젝트/트래블 메이커/Data/scaling_item_table.csv', index_col=0)
item_table = item_table.drop(['리뷰'], axis=1)
item_table.columns = ['item', 'A', 'H', 'F', 'E', 'P', 'S', 'N', 'M']
item_table = item_table[['item', 'H', 'A', 'F', 'E', 'P', 'S', 'N', 'M']]
item_table.head()
```

# 코사인 유사도  
- A와 가장 유사한 거 찾아주는 함수 

```
def similar_users(a):
  user_table_minus_list = {}
  user_table_minus = user_table.drop(a - 1).reset_index().drop(['index'], axis=1)
  user_table_without_USERID = user_table.drop(['USER ID'], axis=1)
  user_table_minus_without_USERID = user_table_minus.drop(['USER ID'], axis=1)
  for i in range(len(user_table_minus)):
    user_table_minus_list[user_table_minus['USER ID'][i]] = np.dot(user_table_without_USERID.loc[a - 1], user_table_minus_without_USERID.loc[i])
  max_users = []
  for keys, values in user_table_minus_list.items():
    if values == max(user_table_minus_list.values()):
      max_users.append('user' + str(keys))
  return max_users
```

```
similar_users(1)
```

# Linear Regression

```
def ratings(b):
  user_table_minus_USERID = user_table.drop(['USER ID'], axis=1)
  item_table_minus_item = item_table.drop(['item'], axis=1)
  items = []
  for i in range(len(item_table)):
    items.append(item_table['item'][i])
  ratings = {}
  for i in range(len(item_table_minus_item)):
    ratings[items[i]] = np.dot(user_table_minus_USERID.iloc[b - 1], item_table_minus_item.iloc[i])
  max_ratings = []
  for keys, values in ratings.items():
    if values == max(ratings.values()):
      max_ratings.append(keys)
  return max_ratings
```

```
ratings(1)
```
