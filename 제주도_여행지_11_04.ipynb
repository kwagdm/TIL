```
df_label = pd.read_csv('/content/drive/My Drive/프로젝트/트래블메이커/제주태그.csv', index_col=0)
```

```
df_activity = df_label[df_label['라벨']=='액티비티']
df_healing = df_label[df_label['라벨']=='휴양']
df_flex = df_label[df_label['라벨']=='플렉스']
df_efficient = df_label[df_label['라벨']=='가성비']
df_private = df_label[df_label['라벨']=='프라이빗']
df_social = df_label[df_label['라벨']=='소셜']

activity = df_activity['태그명'].tolist()
healing = df_healing['태그명'].tolist()
flex = df_flex['태그명'].tolist()
efficient = df_efficient['태그명'].tolist()
private = df_private['태그명'].tolist()
social = df_social['태그명'].tolist()
```

```
for i in range(len(jeju_all)):
  if jeju_all['태그'][i] != 0:
    for word in activity:
      if word in jeju_all['태그'][i]:
        jeju_all['액티비티'][i] += 1
    for word in healing:
      if word in jeju_all['태그'][i]:
        jeju_all['휴양'][i] += 1
    for word in flex:
      if word in jeju_all['태그'][i]:
        jeju_all['플렉스'][i] += 1
    for word in efficient:
      if word in jeju_all['태그'][i]:
        jeju_all['가성비'][i] += 1
    for word in private:
      if word in jeju_all['태그'][i]:
        jeju_all['프라이빗'][i] += 1
    for word in social:
      if word in jeju_all['태그'][i]:
        jeju_all['소셜'][i] += 1
```

```
jeju_no_tagged = jeju_all[jeju_all['태그'] == 0].reset_index().drop(['index'], axis=1)
jeju_no_tagged.head()
```

```
jeju_no_tag = jeju_all[jeju_all['태그'] == 0].index
jeju_tagged = jeju_all.drop(jeju_no_tag).reset_index().drop(['index'], axis=1)
jeju_tagged.head()
```

```
jeju_no_tagged['총갯수'] = 0
for i in range(len(jeju_no_tagged)):
  for j in range(len(jeju_tagged)):
    if jeju_no_tagged['구분'][i] == jeju_tagged['구분'][j]:
      jeju_no_tagged['총갯수'][i] += 1
      jeju_no_tagged['액티비티'][i] += jeju_tagged['액티비티'][j]
      jeju_no_tagged['휴양'][i] += jeju_tagged['휴양'][j]
      jeju_no_tagged['플렉스'][i] += jeju_tagged['플렉스'][j]
      jeju_no_tagged['가성비'][i] += jeju_tagged['가성비'][j]
      jeju_no_tagged['프라이빗'][i] += jeju_tagged['프라이빗'][j]
      jeju_no_tagged['소셜'][i] += jeju_tagged['소셜'][j]
```

```
for i in range(len(jeju_no_tagged)):
  jeju_no_tagged['액티비티'][i] = jeju_no_tagged['액티비티'][i]/jeju_no_tagged['총갯수'][i]
  jeju_no_tagged['휴양'][i] = jeju_no_tagged['휴양'][i]/jeju_no_tagged['총갯수'][i]
  jeju_no_tagged['플렉스'][i] = jeju_no_tagged['플렉스'][i]/jeju_no_tagged['총갯수'][i]
  jeju_no_tagged['가성비'][i] = jeju_no_tagged['가성비'][i]/jeju_no_tagged['총갯수'][i]
  jeju_no_tagged['프라이빗'][i] = jeju_no_tagged['프라이빗'][i]/jeju_no_tagged['총갯수'][i]
  jeju_no_tagged['소셜'][i] = jeju_no_tagged['소셜'][i]/jeju_no_tagged['총갯수'][i]
```
